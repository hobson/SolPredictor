```python
#sns.displot(data=df)
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 7), sharex=False, sharey=False)
axes = axes.ravel()  # array to 1D
cols = df.columns  # [:-1] create a list of dataframe columns to use
cmaps = ['blue', 'green', 'purple', 'pink', 'gray', 'olive']
i=0
for col, ax in zip(cols, axes):
    data = df[[col]]  # select the data
    sns.kdeplot(data=data, x=col, color=cmaps[i], fill=True, ax=ax) #hue='class', cmap=cmaps[i]
    ax.set(title=f'{col} Distribution', xlabel=None)
    i+=1
#fig.delaxes(axes[5])  # delete the empty subplot
fig.tight_layout()
fig.savefig("distributions.jpg", dpi=1200) 
```
